# **TENSOR MATH OPERATIONS**
* Add, Subtract, Divide,Exponents...
* Shorthand and Longhand
* Reassignment

```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#Addition
tensor_a + tensor_b
```

```python
#addition longhand
torch.add(tensor_a, tensor_b)
```

```python
#subtraction
tensor_b - tensor_a
```

```python
#sub function
torch.sub(tensor_b, tensor_a)
```

```python
#multiplication
tensor_a*tensor_b
```

```python
#mul longhand
torch.mul(tensor_a, tensor_b)
```

```python
#division / to float, // to int
tensor_b/tensor_a
```

```python
#division longhand
torch.div(tensor_b, tensor_a)
```

```python
#remainder modulus
tensor_b%tensor_a
```

```python
#remainder longhand
torch.remainder(tensor_b,tensor_a)
```

```python
#Exponents/ pow
tensor_b**tensor_a
torch.pow(tensor_b, tensor_a)
```

```python
# another way to write longhand
tensor_a.add(tensor_b)
```

```python
#Reassignment using underscore
tensor_a +tensor_b
```
